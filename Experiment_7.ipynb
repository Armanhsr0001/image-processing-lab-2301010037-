```python
import cv2
import  numpy as np
```

```python
img=cv2.imread('/content/large_img-worlds-dubai.jpg')
```

```python
row,col=img.shape[:2]
```

```python
M_trans=np.float32([[1,0,50],[0,1,50]])
#50 points shifted
```

```python
translated=cv2.warpAffine(img,M_trans,(col,row))
```

```python
cv2.imwrite("translation.jpg",translated)
```

```python
scaled=cv2.resize(img,None,fx=0.5,fy=0.5)
#fx=width  fy=height
```

```python
cv2.imwrite("scaled.jpg",scaled)
```

```python
M_rot=cv2.getRotationMatrix2D((col/2,row/2),60,1)
#finding img center , 60 digree rotare , 1 scaling
```

```python
rotated=cv2.warpAffine(img,M_rot,(col,row))
```

```python
cv2.imwrite("rotated.jpg",rotated)
```
